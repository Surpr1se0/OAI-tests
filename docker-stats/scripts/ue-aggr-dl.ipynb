# Main Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
from matplotlib.ticker import MultipleLocator
import re
import os

# import data
ul_500 = os.path.join(".." , "clean-files", "500", "agg-500.csv")
ul_200 = os.path.join("..", "clean-files", "200", "agg-200.csv")
ul_100 = os.path.join("..", "clean-files", "100", "agg-100.csv")
ul_50 = os.path.join("..", "clean-files", "50", "agg-50.csv")
ul_10 = os.path.join("..", "clean-files", "10", "agg-10.csv")
ul_5 = os.path.join("..", "clean-files", "5", "agg-5.csv")
ul_1 = os.path.join("..", "clean-files", "1", "agg-1.csv")
```

# Calculations for every file

```python
# Read content of the file
df_ul_1 = pd.read_csv(ul_1)
df_ul_5 = pd.read_csv(ul_5)
df_ul_10 = pd.read_csv(ul_10)
df_ul_50 = pd.read_csv(ul_50)
df_ul_100 = pd.read_csv(ul_100)
df_ul_200 = pd.read_csv(ul_200)
df_ul_500 = pd.read_csv(ul_500)
```

# Behavior Graphics 

## Packet Loss (%)

```python
df_ul_1["Mean_Loss_pct"] = df_ul_1["Mean_Loss_pct"].round(3)
df_ul_5["Mean_Loss_pct"] = df_ul_5["Mean_Loss_pct"].round(3)
df_ul_10["Mean_Loss_pct"] = df_ul_10["Mean_Loss_pct"].round(3)
df_ul_50["Mean_Loss_pct"] = df_ul_50["Mean_Loss_pct"].round(3)
df_ul_100["Mean_Loss_pct"] = df_ul_100["Mean_Loss_pct"].round(3)
df_ul_200["Mean_Loss_pct"] = df_ul_200["Mean_Loss_pct"].round(3)
df_ul_500["Mean_Loss_pct"] = df_ul_500["Mean_Loss_pct"].round(3)

plt.figure(figsize=(14, 6))
sns.lineplot(x="Time_s", y="Mean_Loss_pct", data=df_ul_1, label=" 1Mbps")
sns.lineplot(x="Time_s", y="Mean_Loss_pct", data=df_ul_5, label=" 5Mbps")
sns.lineplot(x="Time_s", y="Mean_Loss_pct", data=df_ul_10, label="  10Mbps")
sns.lineplot(x="Time_s", y="Mean_Loss_pct", data=df_ul_50, label=" 50Mbps")
sns.lineplot(x="Time_s", y="Mean_Loss_pct", data=df_ul_100, label=" 100Mbps")
sns.lineplot(x="Time_s", y="Mean_Loss_pct", data=df_ul_200, label=" 200Mbps")
sns.lineplot(x="Time_s", y="Mean_Loss_pct", data=df_ul_500, label=" 500Mbps")
plt.title("Mean Packet Loss (%) in Uplink\n(Cell Based 10x UE)")
plt.xlabel("Time (ms)")
plt.ylabel("(%)")
plt.legend()
plt.ylim(0, 100)
plt.xlim(left=0)
plt.gca().xaxis.set_major_locator(MultipleLocator(10))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
