# Main Imports

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import defaultdict
from matplotlib.ticker import MultipleLocator
import re
import os

# import data
cb_10x = os.path.join("..", "clean-files-dl", "aggr", "agg-cb-10.csv")
cf_10x = os.path.join("..", "clean-files-dl", "aggr", "agg-cf-10.csv")
cb_5x = os.path.join("..", "clean-files-dl", "aggr", "agg-cb-5.csv")
cf_5x = os.path.join("..", "clean-files-dl", "aggr", "agg-cf-5.csv")
cb_1x = os.path.join("..", "clean-files-dl", "aggr", "agg-cb-1.csv")
cf_1x = os.path.join("..", "clean-files-dl", "aggr", "agg-cf-1.csv")
```

# Calculations for every file

#### 1x Scenarios

```python
# Read content of the file
df_cb_1 = pd.read_csv(cb_1x)
df_cf_1 = pd.read_csv(cf_1x)
```

#### 5x Scenarios

```python
def read_data_by_ue(csv_path): 
  df = pd.read_csv(csv_path)

  dfs = {}

  for ue_id, group in df.groupby("UE_ID"):
    df_ue = group.sort_values("Time_ms").reset_index(drop=True)
    df_ue["Rolling Avg (Mbps)"] = df_ue["Mean_Throughput"].rolling(window=5, min_periods=1).mean()
    dfs[ue_id] = df_ue
  
  return dfs

dfs_cb_5 = read_data_by_ue(cb_5x)
dfs_cf_5 = read_data_by_ue(cf_5x)
```

#### 10x Scenarios

```python
def read_data_by_ue(csv_path): 
  df = pd.read_csv(csv_path)

  dfs = {}

  for ue_id, group in df.groupby("UE_ID"):
    df_ue = group.sort_values("Time_ms").reset_index(drop=True)
    df_ue["Rolling Avg (Mbps)"] = df_ue["Mean_Throughput"].rolling(window=5, min_periods=1).mean()
    dfs[ue_id] = df_ue
  
  return dfs

dfs_cb_10 = read_data_by_ue(cb_10x)
dfs_cf_10 = read_data_by_ue(cf_10x)
```

# Behavior Graphics 

## 1X Scenarios

```python
# Round of throughput values to 3 decimal places
df_cb_1["Mean_Throughput"] = df_cb_1["Mean_Throughput"].round(3)
# Calculate rolling average with a window of 5
df_cb_1["Rolling Avg (Mbps)"] = df_cb_1["Mean_Throughput"].rolling(window=5).mean()

# Graphic of Throughput and Rolling Average
plt.figure(figsize=(14, 6))
sns.lineplot(x="Time_ms", y="Mean_Throughput", data=df_cb_1, label="Mean Throughput Downlink (mbps)")
sns.lineplot(x="Time_ms", y="Rolling Avg (Mbps)", data=df_cb_1, label="Rolling Avg")
plt.title("Mean Throughput in time for Downlink\n(Cell Based 1x UE)")
plt.xlabel("Time (ms)")
plt.ylabel("Throughput Downlink(Mbps)")
plt.legend()
plt.ylim(0, 2)
plt.xlim(left=0)
plt.gca().xaxis.set_major_locator(MultipleLocator(5000))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()


# Round of throughput values to 3 decimal places
df_cf_1["Mean_Throughput"] = df_cf_1["Mean_Throughput"].round(3)
# Calculate rolling average with a window of 5
df_cf_1["Rolling Avg (Mbps)"] = df_cf_1["Mean_Throughput"].rolling(window=5).mean()

# Graphic of Throughput and Rolling Average
plt.figure(figsize=(14, 6))
sns.lineplot(x="Time_ms", y="Mean_Throughput", data=df_cf_1, label="Mean Throughput Downlink (mbps)")
sns.lineplot(x="Time_ms", y="Rolling Avg (Mbps)", data=df_cf_1, label="Rolling Avg")
plt.title("Mean Throughput in time for Downlink\n(Cell Free 1x UE)")
plt.xlabel("Time (ms)")
plt.ylabel("Throughput Downlink(Mbps)")
plt.legend()

plt.ylim(0, 2)
plt.xlim(left=0)
plt.gca().xaxis.set_major_locator(MultipleLocator(5000))
plt.grid(True, linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```
